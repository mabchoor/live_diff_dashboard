```python
# compare_binance_okx_live.py
# Real-time Binance vs OKX USDT prices (auto-refresh)

import requests
import time
import os
from rich.console import Console
from rich.table import Table
from datetime import datetime

console = Console()

def fetch_binance_tickers():
    urls = [
        "https://data-api.binance.vision/api/v3/ticker/price",
        "https://data-api.binance.vision/api/v3/ticker/price"
    ]
    for url in urls:
        try:
            r = requests.get(url, timeout=15)
            r.raise_for_status()
            data = r.json()
            out = {it["symbol"]: float(it["price"]) for it in data if it["symbol"].endswith("USDT")}
            return out
        except Exception as e:
            console.print(f"[yellow]âš ï¸ Binance fetch failed ({url}): {e}[/yellow]")
    return {}

def fetch_okx_tickers():
    url = "https://www.okx.com/api/v5/market/tickers?instType=SPOT"
    try:
        r = requests.get(url, timeout=15)
        r.raise_for_status()
        data = r.json()
        return {it["instId"].upper(): float(it["last"]) for it in data.get("data", []) if it["instId"].endswith("-USDT")}
    except Exception as e:
        console.print(f"[yellow]âš ï¸ OKX fetch failed: {e}[/yellow]")
        return {}

def normalize_binance_symbol(sym):
    return f"{sym[:-4]}-USDT" if sym.endswith("USDT") else sym

def compute_differences(binance, okx):
    results = []
    bnorm = {normalize_binance_symbol(s): p for s, p in binance.items()}
    common = set(bnorm.keys()) & set(okx.keys())
    for ns in common:
        bprice = bnorm[ns]
        oprice = okx[ns]
        abs_diff = abs(bprice - oprice)
        avg = (bprice + oprice) / 2
        pct_diff = (abs_diff / avg) * 100
        results.append((ns, bprice, oprice, abs_diff, pct_diff))
    return sorted(results, key=lambda x: x[4], reverse=True)

def main():
    refresh_interval = 10  # seconds
    console.print("[bold green]ğŸ”„ Starting real-time Binance vs OKX monitor...[/bold green]")

    while True:
        binance = fetch_binance_tickers()
        okx = fetch_okx_tickers()

        if not binance or not okx:
            console.print("[red]âŒ Missing data from one of the exchanges.[/red]")
            time.sleep(refresh_interval)
            continue

        diffs = compute_differences(binance, okx)
        top_n = 10

        table = Table(show_header=True, header_style="bold magenta")
        table.add_column("Pair", justify="left")
        table.add_column("Binance", justify="right")
        table.add_column("OKX", justify="right")
        table.add_column("AbsDiff", justify="right")
        table.add_column("%Diff", justify="right")

        for (pair, b, o, diff, pct) in diffs[:top_n]:
            color = "green" if b > o else "red"
            table.add_row(pair, f"{b:.6f}", f"{o:.6f}", f"{diff:.6f}", f"[{color}]{pct:.3f}%[/{color}]")

        os.system("clear" if os.name != "nt" else "cls")
        console.print(f"[cyan]Last update: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}[/cyan]")
        console.print(table)
        console.print(f"Next update in {refresh_interval}s...\n")
        time.sleep(refresh_interval)

if __name__ == "__main__":
    main()
```

```python
import requests
import pandas as pd
from IPython.display import display, clear_output
from time import sleep
from datetime import datetime

def fetch_binance_tickers():
    url = "https://data-api.binance.vision/api/v3/ticker/price"
    r = requests.get(url, timeout=10)
    r.raise_for_status()
    data = r.json()
    return {it["symbol"]: float(it["price"]) for it in data if it["symbol"].endswith("USDT")}

def fetch_okx_tickers():
    url = "https://www.okx.com/api/v5/market/tickers?instType=SPOT"
    r = requests.get(url, timeout=10)
    r.raise_for_status()
    data = r.json()
    return {it["instId"].upper(): float(it["last"]) for it in data.get("data", []) if it["instId"].endswith("-USDT")}

def normalize_binance_symbol(sym):
    return f"{sym[:-4]}-USDT"

def compute_differences(binance, okx):
    results = []
    bnorm = {normalize_binance_symbol(s): p for s, p in binance.items()}
    common = set(bnorm.keys()) & set(okx.keys())

    for ns in common:
        bprice = bnorm[ns]
        oprice = okx[ns]
        if not bprice or not oprice or bprice == 0 or oprice == 0:
            continue
        abs_diff = abs(bprice - oprice)
        avg = (bprice + oprice) / 2
        pct_diff = (abs_diff / avg) * 100
        if pct_diff != 0:
            results.append((ns, bprice, oprice, abs_diff, pct_diff))
    return sorted(results, key=lambda x: x[4], reverse=True)

def live_refresh(interval=30, top_n=20):
    try:
        while True:
            binance = fetch_binance_tickers()
            okx = fetch_okx_tickers()
            diffs = compute_differences(binance, okx)

            # Build DataFrame for display
            df = pd.DataFrame(diffs, columns=["Pair", "Binance Price", "OKX Price", "Abs Diff", "% Diff"])
            df_display = df.head(top_n).copy()
            df_display["Binance Price"] = df_display["Binance Price"].map("{:.6f}".format)
            df_display["OKX Price"] = df_display["OKX Price"].map("{:.6f}".format)
            df_display["Abs Diff"] = df_display["Abs Diff"].map("{:.6f}".format)
            df_display["% Diff"] = df_display["% Diff"].map("{:.4f}%".format)

            clear_output(wait=True)
            print(f"Last updated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
            print(f"Showing top {top_n} price differences between Binance and OKX:\n")
            display(df_display)

            sleep(interval)

    except KeyboardInterrupt:
        print("Live refresh stopped by user.")

# Run it (interrupt cell execution to stop)
live_refresh(interval=30, top_n=20)
```
